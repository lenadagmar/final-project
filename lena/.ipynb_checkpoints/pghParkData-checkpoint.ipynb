## The Prettiest Neighborhood: Amount of Parks/Green Space

```python
import pandas as pd
parks = pd.read_csv("pghParks.csv")
parks[['name','type','neighborhood']]

#each park in pittsburgh (209 total)
```

This database is a list of all 209 parks in Pittsburgh, and what neighborhood they are located in. 

```python
nbhdParks={}

for nbhd in parks['neighborhood']:
    parkCount=0
    for park in range(len(parks)):
        if parks.iloc[park, 6]==nbhd:
            parkCount+=1
        nbhdParks[nbhd]=parkCount
    if nbhd not in nbhdParks:
        nbhdParks[nbhd]=0
    
nbhds=pd.DataFrame.from_dict(nbhdParks, orient='index')
nbhds=nbhds.sort_values(0, ascending=False)
nbhds.head(100)

#number of parks in each neighborhood, descending order
```

 This table lists the number of parks per neighborhood in Pittsburgh. 

```python
nbhds.describe()
```

```python
list = parks['type'].unique()
for item in list:
    print(item)
```

Each park falls into one of the above 5 categories, however, we will only focus on the type "Park"

```python
parkSeries = pd.Series(parks['type'])
parkQueryMask = parkSeries.str.contains("Park")
parkTypes = parks[parkQueryMask]
parkTypes.head(5)
```

```python
nbhdJustParks={}

for nbhd in parkTypes['neighborhood']:
    parkCount=0
    for park in range(len(parkTypes)):
        if parks.iloc[park, 6]==nbhd:
            parkCount+=1
        nbhdParks[nbhd]=parkCount
    if nbhd not in nbhdParks:
        nbhdParks[nbhd]=0
    
nbhdJustParks=pd.DataFrame.from_dict(nbhdParks, orient='index')
nbhdJustParks=nbhdJustParks.sort_values(0, ascending=False)
new="# of parks"
nbhdJustParks.index.rename(new, inplace=True)

nbhdJustParks.head(30)

#number of parks in each neighborhood, excluding traffic medians, building grounds, monuments and open spaces, descending order
```

This table represents the number of parks in each neighborhood, excluding traffic medians, building grounds, monuments and open spaces.

```python
nbhdJustParks.describe()
```

Since the average amount of parks per neighborhood is 2, we will eliminate any neighborhood with 2 or less parks.                        

```python
nbhdJustParks=(nbhdJustParks.loc[nbhdJustParks[0] > 2])
nbhdJustParks.head(10)
```
