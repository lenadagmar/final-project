As Pittsburgh is the city of bridges, we will be looking at the various different bridges in all of the neighborhoods. Our ideal neighborhood has as many bridges as possible because bridges are a core part of the city's history. 

```python
import pandas as pd
bridges = pd.read_csv("bridges.csv")
bridges[['name', 'neighborhood']]
```

This database is a list of a portion of the bridges in Pittsburgh and what neighborhood they are located in.

```python
nbhdBridges = {}

for nbhd in bridges['neighborhood']:
    bridgeCount = 0
    for bridge in range(len(bridges)):
        if bridges.iloc[bridge, 8] == nbhd:
            bridgeCount += 1
        nbhdBridges[nbhd] = bridgeCount
    if nbhd not in nbhdBridges:
        nbhdBridges[nbhd] = 0
    
nbhds = pd.DataFrame.from_dict(nbhdBridges, orient = 'index')
nbhds = nbhds.sort_values(0, ascending = False)
new= "number of bridges"
nbhds.index.rename(new, inplace = True)
nbhds.head(100)

#number of parks in each neighborhood, descending order
```

This table lists the number of bridges per neighborhood in Pittsburgh.

```python
nbhds.describe()
```

Since the mean is 2.8, we will elimate all the neighborhoods with less than 3 parks.

```python
nbhds = (nbhds.loc[nbhds[0] > 3])
nbhds.head(10)
```

```python
nbhds[:10].plot(kind = 'bar', title = "Number of Bridges")
```
